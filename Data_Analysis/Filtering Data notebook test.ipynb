```python
def filterDataForPipline(data):
    filteredData = []
    for i in range(len(data[0])):
        filterData = [0, []]
        if 'Face' in data[0][i].get('markers').get('data'):
            filterData[0] = 1
        filterData[1] = data[0][1].get('eeg').get('data')
        filteredData.append(filterData)
    return filteredData
```

```python
def filterDataForPipline2(data):
    trialData = getMarkerBoundSingleMarkerData('NewImageStart', 'NewImageEnd', data, go_backward=False)
    filteredData = []
    #print(len(trialData[0]))
    for i in range(len(trialData[0])):
        filterData = [0, []]
        if 'Face' in trialData[0][i].get('markers').get('data'):
            filterData[0] = 1
        filterData[1] = trialData[0][1].get('eeg').get('data')
        filteredData.append(filterData)
        
    return filteredData
```
